```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
store_data = pd.read_csv("GroceryStoreDataSet.csv")
store_data.head(7)
```

```python
# Inisialisasi list kosong untuk menyimpan transaksi
records = []

# Loop untuk membaca setiap baris dalam dataset
for i in range(0, store_data.shape[0]):
    # Ambil nilai pada kolom pertama, pisahkan item berdasarkan koma,
    # hapus spasi berlebih, lalu tambahkan ke list 'records'
    records.append([str(item).strip() for item in store_data.values[i, 0].split(',')])

# Cetak hasil akhir dalam bentuk list of lists
print(records)
```

```python
# Menerapkan algoritma Apriori pada data transaksi (records)
# Parameter:
# - min_support = 0.003: Nilai minimum support untuk itemset yang dipertimbangkan
# - min_confidence = 0.2: Nilai minimum confidence untuk aturan asosiasi
# - min_lift = 3: Nilai minimum lift untuk menilai kualitas aturan
# - min_length = 2: Panjang minimum itemset yang dianalisis
association_rules = apriori(records, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2)

# Mengonversi hasil algoritma Apriori menjadi list
association_results = list(association_rules)

# Loop untuk menampilkan setiap aturan asosiasi yang ditemukan
for i in association_results:
    print(i, "\n")  # Cetak hasil setiap aturan asosiasi dalam format RelationRecord
```

```python
print(association_results[0])
```

```python
# Loop untuk menampilkan aturan asosiasi dalam format yang lebih mudah dibaca
for item in association_results:
    # Mengambil itemset dari hasil aturan asosiasi (pair adalah frozenset item-item)
    pair = item[0]
    # Mengonversi frozenset menjadi list untuk memudahkan akses item individual
    items = [x for x in pair]

    # Menampilkan aturan asosiasi dalam format "A -> B"
    print("Rule: " + items[0] + " -> " + items[1])
    # Menampilkan nilai support untuk aturan asosiasi
    print("Support: " + str(item[1]))
    # Menampilkan nilai confidence dari aturan asosiasi pertama dalam ordered_statistics
    print("Confidence: " + str(item[2][0][2]))
    # Menampilkan nilai lift untuk menilai kualitas aturan asosiasi
    print("Lift: " + str(item[2][0][3]))
    # Membuat pemisah untuk setiap aturan asosiasi
    print("========================================")
```

```python
# Menerapkan algoritma Apriori pada data transaksi (records)
# Parameter:
# - min_support = 0.007: Nilai minimum support yang lebih tinggi dibandingkan sebelumnya,
#   untuk memfilter itemset dengan frekuensi lebih signifikan
# - min_confidence = 0.6: Minimum confidence dinaikkan untuk memastikan aturan dengan kepercayaan tinggi
# - min_lift = 3: Minimum lift tetap 3 untuk menjaga kualitas aturan asosiasi
# - min_length = 2: Panjang minimum itemset tetap 2

association_rules = apriori(records, min_support=0.007, min_confidence=0.6, min_lift=3, min_length=2)

# Mengonversi hasil algoritma Apriori menjadi list
association_results = list(association_rules)

# Loop untuk menampilkan setiap aturan asosiasi yang ditemukan
for i in association_results:
    print(i, "\n")  # Cetak aturan asosiasi dalam format RelationRecord
```

```python
print(association_results[0])
```

```python
# Loop untuk menampilkan aturan asosiasi yang ditemukan dalam format yang lebih mudah dibaca
for item in association_results:
    # Mengambil itemset dari aturan asosiasi (item[0] adalah frozenset yang berisi item-item dalam aturan)
    pair = item[0]
    # Mengonversi frozenset menjadi list untuk memudahkan akses setiap item
    items = [x for x in pair]

    # Menampilkan aturan asosiasi dalam format "A -> B"
    print("Rule: " + items[0] + " -> " + items[1])
    # Menampilkan nilai support, yaitu proporsi transaksi yang mengandung itemset
    print("Support: " + str(item[1]))
    # Menampilkan nilai confidence, yaitu probabilitas kemunculan item kedua saat item pertama muncul
    print("Confidence: " + str(item[2][0][2]))
    # Menampilkan nilai lift, yaitu pengukuran kualitas aturan; lift > 1 menunjukkan hubungan yang signifikan
    print("Lift: " + str(item[2][0][3]))
    # Menampilkan pemisah untuk memisahkan setiap aturan asosiasi
    print("========================================")
```

Minimum support menentukan frekuensi minimum itemset yang dipertimbangkan, minimum confidence memastikan kekuatan hubungan antar item dalam aturan, sementara lift mengevaluasi seberapa signifikan hubungan tersebut dibandingkan dengan kemunculan independen item.
